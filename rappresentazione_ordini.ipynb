**`Rappresentazione ordini con deviazione standard nota`**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
numero_estrazioni = 200
### rappresentazione dei dati con deviazione standard nota
ordini_lunedì = np.random.normal(250,2,size=numero_estrazioni)
ordini_martedì = np.random.normal(150,1,size=numero_estrazioni)
ordini_mercoledì = np.random.normal(90,2,size=numero_estrazioni)
ordini_giovedì = np.random.normal(80,2,size=numero_estrazioni)
ordini_venerdì = np.random.normal(100,2,size=numero_estrazioni)
plt.hist(ordini_lunedì)
plt.title(f'Distribuzione degli ordini per ogni lunedì (numero lunedi = {numero_estrazioni})')
plt.show()
plt.hist(ordini_martedì)
plt.title(f'Distribuzione degli ordini per ogni martedì (numero martedì = {numero_estrazioni})')
plt.show()
plt.hist(ordini_mercoledì)
plt.title(f'Distribuzione degli ordini per ogni mercoledì (numero mercoledì = {numero_estrazioni})')
plt.show()
plt.hist(ordini_giovedì)
plt.title(f'Distribuzione degli ordini per ogni giovedì (numero giovedì = {numero_estrazioni})')
plt.show()
plt.hist(ordini_venerdì)
plt.title(f'Distribuzione degli ordini per ogni venerdì (numero venerdì = {numero_estrazioni})')
plt.show()
array=np.empty([numero_estrazioni,5])

array[:,0]=ordini_lunedì
array[:,1]=ordini_martedì
array[:,2]=ordini_mercoledì
array[:,3]=ordini_giovedì
array[:,4]=ordini_venerdì

df=pd.DataFrame(array)
df.columns=['lunedì','martedì','mercoledì','giovedì','venerdì']

sns.boxenplot(data=df)
plt.title(f'rappresentazione per diffusione')
plt.show()
df

```

```python
def funzione(a,b,c,d,e):
  return 0.1*ordini_lunedì+3*ordini_martedì+0.1+ordini_mercoledì**2+0.01*ordini_giovedì*ordini_venerdì

target=funzione(ordini_lunedì,ordini_martedì,ordini_mercoledì,ordini_giovedì,ordini_venerdì)
df['target']=target
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x=df.iloc[:,0:5]
y=df.iloc[:,5]

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25)

lreg=LinearRegression()

lreg.fit(X_train,y_train)

y_pred=lreg.predict(X_test)

score = lreg.score(X_test,y_test)
print('lo score sui dati di test: '+str(score))

score = lreg.score(X_train,y_train)
print('lo score sui dati di train: '+str(score))

plt.scatter(y_pred,y_test)
plt.plot(y_pred,y_pred,color='red')
plt.show()
```

```python
import random
import numpy as np


# creiamo una lista con i giorni della settimana, escluso sabato e domenica
giorni_settimana = ["lunedi", "martedi", "mercoledi", "giovedi", "venerdi"]

# creiamo una lista vuota che conterrà i risultati
risultati = []

# ripetiamo l'operazione 1000 volte
for i in range(1000):
    # creiamo una lista vuota per i numeri casuali di questa iterazione
    numeri_casuali = []
    # generiamo un numero casuale per ogni giorno della settimana
    for giorno in giorni_settimana:
        if giorno == 'lunedi':
            numero_casuale = random.randint(250, 250)
        elif giorno == 'martedi':
            numero_casuale = random.randint(200, 200)
        elif giorno == 'mercoledi':
            numero_casuale = random.randint(150, 150)
        elif giorno == 'giovedi':
            numero_casuale = random.randint(120, 120) 
        elif giorno == 'venerdi':
            numero_casuale = random.randint(200, 200) 
        else:
          continue                      
        numeri_casuali.append(numero_casuale)
    # aggiungiamo i numeri casuali di questa iterazione alla lista dei risultati
    risultati.append(numeri_casuali)

# trasponiamo la lista dei risultati per ottenere un array con 1000 righe e 5 colonne
risultati_array = [[risultati[j][i] for j in range(len(risultati))] for i in range(len(risultati[0]))]

# stampiamo l'array risultati_array
for riga in risultati_array:
    X = riga
    X_medio = np.mean(X)
    SD = np.std(X, ddof=1)
    print(SD)
```
